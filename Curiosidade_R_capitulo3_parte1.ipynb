# Avaliar a estacionaridade ou estabilidade de uma série temporal

Como o comportamento de longo prazo do passado reflete no futuro ao longo prazo. Verificar a dinâmica da série temporal (como mudanças sazonais por exemplo). 

Fazer a "self-correlations" para responder à pergunta: com que precisão os dados passados, distantes ou recentes, predizem os dados dos futuros?

Depois ver as "correlações espúrias": para ter certeza que não estamos identificando relações baseadas em dinâmicas que não envolvem de forma alguma as relações causais que queremos descobrir. 

### Estacionário
Muitos modelos estatísticos dependem de uma série temporal estacionária (estáveis ao longo do tempo no que diz respeito ).

Não é estacioonário: 
- o valor médio está aumentando com o tempo, em vez de estar estável.
- A distância entre as oscilações pico-vale em uma base anual está crescendo
- forte comportamento sazonal, sendo oposto a estacionaridade

##### Teste de estacionaridade e o teste Dickey-Fuller aumentado

Processo é estacionário para todos os possíveis lags k, a distribuição de yt, yt+1,.., yt+k não depender de t. 
Saber se existe uma raiz unitária - se 1 é uma solução da equação característica do processo. Uma série temporal linear não é estacionária se houver uma raiz unitária, embora uma raiz unitária não prove a estacionaridade. 
Ainda segue sendo pesquisado se um processo tem uma raiz unitária.


Raiz unitária => Passeio aleatório (random walk) 

y_t = ϕ × y_t-1 + e_t

ϕ = 1 , a série tem raiz unitária. Ela "fugirá" e não será estacionária. 
Dickey-Fuller aumentado (ADF) 
Mesmo a série não sendo estacionária, isso não significa que ela deva ter uma tendência. Exemplo de passeio e preços de ações.


Verificar se o processo é estacionário se chamam "testes de hipóteses". Teste Dickey-Fuller aumentado (ADF) é a métrica comumente usada para avaliar uma série temporal quanto se trata de problemas de estacionariedade. É a hipótese nula de que uma raiz unitária está presente em uma série temporal. Uma série integrada é estacionária. 
Uma série integrada de ordem d é uma série que deve ser diferenciada d vezes para se tornar estacionária.  A estrutura do teste Dickey-Fuller é:

Δy_t = y_t – y_t-1 = (ϕ – 1) × y_t-1 + ϵ_t

Então, o teste para saber se ϕ=1 é um simples teste t que verifica se o parâmetro no lag y_t-1 é igual a zero. A diferença é que o teste ADF leva em consideração mais lags, enquanto o modelo sujacente leva em consideração a dinâmica de ordem superior, que pode ser escrita como série de lags diferenciados.


Y_t – ϕ_1 × y_t-1 – ϕ_2 × y_t-2 . . . = ϵ_t


Dickey-Fuller aumentado (ADF) | Dickey-Fuller aumentado (ADF) =>  O teste ADF é o teste de estacionariedade usado com a maior frequência nos estudos e pesquisas de séries temporais. Mas não são milagrosos.
-  não conseguem distinguir muito as raízes unitárias próximas de ráizes unitárias.
-  já que o tamanho amostral é reduzido, falsos positivos para raízes unitárias são bastantes comuns


série temporal pode se tornar estacionária, com transformação log e uma transformação raiz quadrada, no caso de mudança da variância ao longo do tempo. A remoção de uma tendência é normalmente feita por diferenciação. Mas, o ideal não é diferenciar por muitas vezes, isso é improvável resolver o problema.

-- Normalização da distribuição das variáveis de entrada ou variável predita.

Transformação Box Cox, implementada no R pelo pacote forecast e no python pelo pacorte scipy.stats.

No entanto, às vezes um alinhamento à esquerda pode ser útil para fazer perguntas exploratórias, como "se eu soubesse disso com antecedência, me ajudaria?". Mas, geralmente nem um lookahead informa muita coisa, o que a variável não é informativa.
```python
rolla_r= """


# data is a zoo object or a plain vector or matrix
#
# width is 
# - a list of integer vectors representing offsets or a plain vector of widths.
#   There is one per time point or its recycled if too short.  recycling uses
#   by= argument if length(width) is 1; otherwise, by is ignored.
#   If width represents widths then they are turned into offsets using align.
#
# If we are at 5th time of data and width[[5]] is c(-2,-1,0) then FUN is applied
#   to positions i + width[[i]] = 5 + c(-2,-1,0)  = 3:5 of z (so in terms of
#   a width specification it would be the same as width = 3, align = "right").
#
# Therefore we have the following transformations:
#   widths are converted to offsets which are converted to positions.  
#   The offsets are the components of width and the 
#   positions are i+width[[i]] after partial processing.  partial can be:
#   - logical.  FALSE means that all offets must exist or else no result is 
#     produced for that time point.  TRUE means that at least one offset must 
#     exist.
#   - numeric.  The minimum number of offsets that must exist.  If < 0 then
#     all elements of the offset must exist.  Note that TRUE corresponds to 1
#     and FALSE correspoinds to -1.  These are the two most common values.
#
# For points that are not computed they are filled in with fill.  fill has 
#   three elements and is recycled if too short.   fill = NULL is the default.
#   The elements represent what to fill the left points, interior points and
#   right points.  NULL causes no filling and "extend" causes the first or
#   last point to be repeated or interior points to be linearly approximated.

# wrapper around rollapply which defaults to align = "right"
rollapplyr <- function(..., align = "right") {
	rollapply(..., align = align)
}

rollapply <- function(data, ...) UseMethod("rollapply")

rollapply.default <- function(data, ...) {
        if (length(data) < 1L) return(data)
	coredata(rollapply(zoo(data), ...))
}

rollapply.ts <- function(data, ...) {
        if (length(data) < 1L) return(data)
	as.ts(rollapply(as.zoo(data), ...))
}

rollapply.zoo <- function(data, width, FUN, ..., by = 1, 
	by.column = TRUE, fill = if (na.pad) NA, na.pad = FALSE,
	partial = FALSE, align = c("center", "left", "right"), coredata = TRUE) {

        if (length(data) < 1L) return(data)

	if (!missing(na.pad)) {
		warning("na.pad argument is deprecated")
	}

    if (is.vector(width) && !is.list(width) && length(width) == 1 &&
		by.column && length(by) == 1 && by == 1 && (missing(partial) | identical(partial, FALSE)) &&
		length(list(...)) < 1 && length(sw <- deparse(substitute(FUN))) == 1) {
		  if (sw == "mean" && !anyNA(data)) {
				return(rollmean(data, width, fill = fill, align = align))
		  } else if (sw == "median" && width %% 2 == 1 && !anyNA(data)) {
				return(rollmedian(data, width, fill = fill, align = align))
	      } else if (sw == "max") {
				return(rollmax(data, width, fill = fill, align = align))
	      }
	}
	FUN <- match.fun(FUN)

	if (by.column && length(dim(data)) == 2) {
		z <- do.call(merge,
			lapply(1:NCOL(data), function(j)
				rollapply(data[, j, drop = TRUE], width = width, FUN = FUN, ...,
					by = by, by.column = by.column, fill = fill,
					partial = partial, align = align, coredata = coredata)
			)
		)
		if (NCOL(data) == 1) dim(z) <- c(length(z), 1)
		colnames(z) <- if (NCOL(z) == NCOL(data)) colnames(data)
		return(z)
	}

	if (is.logical(partial)) partial <- if (partial) 1 else -1

	# convert widths to offsets using align
	align <- match.arg(align)

	width <- if (!is.list(width)) {
	  lapply(trunc(width), function(w) {
			if (align == "right") seq(to = 0, length.out = w)
			else if (align == "center") seq(to = floor(w/2), length.out = w)
			else seq(from = 0, length.out = w)
	  })
	} else {
	  lapply(width, function(w) {
	                if(is.null(w)) NULL else trunc(w)
	  })
	}
	# recycle width (using by if length(width) == 1)
	width <- if (length(width) == 1) {
		w <- rep(list(NULL), NROW(data))
		start.at <- if (partial < 0) max(-min(width[[1]]), 0) + 1 else 1
		start.at <- min(NROW(data), start.at)
		replace(w, seq(start.at, NROW(data), by = by), width)
	} else rep(width, length.out = NROW(data))

	f <- if (is.null(dim(data))) {
		# undimensioned
		#
		# if FUN is to be evaluated at offsets for the ith point then calculate
		# positions, do partial processing and apply FUN
		function(i, offsets, data, ...) { 
			if (is.null(offsets)) return(NULL)
			posns <- i + offsets
			ix <- posns >= 1 & posns <= NROW(data)
			if (partial < 0) {
				if (all(ix)) FUN(data[posns], ...)
			} else if (sum(ix) >= partial) {
				FUN(data[replace(posns, !ix, 0)], ...)
			}
		}
    } else {
		# dimensioned
		#
		# same f as in TRUE leg except data[.] becomes data[.,]
		function(i, offsets, data, ...) { 
			if (is.null(offsets)) return(NULL)
			posns <- i + offsets
			ix <- posns >= 1 & posns <= NROW(data)
			if (partial < 0) {
				if (all(ix)) FUN(data[posns,], ...)
			} else if (sum(ix) >= partial) {
				FUN(data[replace(posns, !ix, 0),], ...)
			}
		}
	}

        dat <- if(coredata) coredata(data) else data
	dat <- mapply(f, seq_along(time(data)), width, 
		MoreArgs = list(data = dat, ...), SIMPLIFY = FALSE) 
		
	ix <- !sapply(dat, is.null) # integer indexes of non-nulls

	if (!missing(fill) || !missing(na.pad)) {

		# replace NULLs with NAs
		dat <- lapply(dat, function(x) if (is.null(x)) NA else x)

		# construct zoo object
		dat <-
		if (max(sapply(dat, length)) > 1)
			zoo(do.call("rbind", dat), index(data), attr(data, "frequency"))
		else
			zoo(do.call("c", dat), index(data), attr(data, "frequency"))

		# perform filling
		dat <- na.fill(dat, fill, ix)

	} else {

		# construct zoo object removing points corresponding to NULL
		dat <- if (max(sapply(dat, length)) > 1)
			zoo(do.call("rbind", dat), index(data)[ix], attr(data, "frequency"))
		else
			zoo(do.call("c", dat), index(data)[ix], attr(data, "frequency"))
	}

	dat
}
"""
robjects.r(rolla_r)
```

# janela de expansão Expanding windows

São mais restritos, estimando uma síntese estatística que acredita ser um processo estável, em vez de evoluir com o tempo ou oscilar significativamente. Janela de expnação começa com um determinado tamanho mínimo e vai expandindo conforme avança as séries tempraois, incluindo todos os pontos até determinado tempo. 

A função no R de janelas de expansão são cummax e cummin. Dá para fazer uma média cumulativa, com o comando cumsum



```python
import rpy2.robjects as robjects

codigo5="""

#R
# janela de expansão 

plot(x, type ='l', lwd=1)
# max
lines(cummax(x),  col =2, lwd=3, lty=2)
# mean
lines(cumsum(x)/1:length(x), col =3, lwd=5, lty=3)


#tracejado lty
#lwd grossura da linha


"""
robjects.r(codigo5)
```

uma janela de expansão "max" (traços longos e vermelhos e uma janela de expansão "mean" (traços curtos verde). O uso de uma janela de expansão significa que o max sempre reflete o máiximo global até aquele tempo, tornando-a uma função monotônica. Graças à extensa "memória" da janela de expansão, sua média é menor do que a média móvel (figura anterior), por que a tendência subjacenge é menos proeminente na média de expansão. # Identificando a Self-correlation


Autocorrelação espacial ou autocorrelação de valores de um dado atributo originado de localizações relativas no espaço geográfico. 


Temperaturas de 15/05 de vários anos, com as temperaturas 15/08 de vários anos. Há  correlação com as máximas de 15/05 e de 15/08, e as mínimas de 15/05 e 15/08. 
Correlação próxima de zero, saber a temperatura de 15/05 não fornece as informações prováveis da variação da temperatura no dia 15/08. A grosso modo isso é self-correlation.


A autocorrelação generaliza a self-correlation, não se aconra em um ponto específico no tempo. Existir correlação entre quaisquer dois pontos. 

Autocorrelação dá uma ideia de como os pontos de dados em diferentes pontos no tempo estão linearmente relacionados entre si em função de sua diferença de tempo. 

A <i> função de autocorrelação (ACF) </i> pode ser compreendida intuitivamente como uma representação gráfica. Plotando em R. 

```python
#outra maneira de plotar este gráfico 

import rpy2.robjects as robjects

codigo5 = """
x <- 1:100

y <- sin(x * pi/3)

plot(y, type= 'b')
acf(y)
"""

robjects.r(codigo5)
```
