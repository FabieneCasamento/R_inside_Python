## Correlações Espúrias


Sazonalidade - pense em uma correlação espúria entre o consumo de cachorro-quente e a morte por afogamento (verão). 


Mudança de nível ou declive nos dados de mudança ao longo do tempo (gerando uma distribuição dumbbell [dumbbell dot plots ou gráfico halteres] com alta correlação sem sentido). 


Quantidades somada cumulativas (este é um truque usado em determinados setores para fazer com que os modelos ou correlações pareçam melhores do que são). 


Cointergração tem alta correlações. 

A diferença ente correlação espúria e a cointegração é que na espúria não precisa haver nenhuma relação, ao passo que as séries temporais cointegradas estão fortemente relacionadas entre si. 


Conseguimos identificar a correlação espúria?

 <b>Correlação</b> é qualquer dependência ou associação (casual ou não) entre duas variáveis. Este conceito é uma medida matemática que busca entender a força da relação entre as diversas dimensões.

Já palavra <b>espúria</b> pode ser entendida nesse contexto como algo que não é verdadeiro, real ou hipotético.


O número de afogamentos relacionado a quantidade de filmes em que o Nicolas Cage aparece por ano. Essencialmente o caso acima é absurdo, ilustrativo e didático.

```python
cod = """ # Define the hist2d function
hist2d <- function(data, nbins.y, xlabels){
    # Fazemos ybins com espaçamento uniforme
    # para incluir pontos mínimos e máximos
    ymin = min(data)
    ymax = max(data) * 1.0001
    # saída lazy para evitar a preocupação com inclusão/ exclusão

    ybins = seq(from = ymin, to = ymax, length.out = nbins.y + 1)

    # Faz uma matriz do zero do tamanho apropriado
    hist.matrix = matrix(0, nrow = nbins.y, ncol = ncol(data))

    ## Os dados vêm em forma de matriz, 
    ## em que cada linha representa um ponto de dados.

    for (i in 1:nrow(data)){
        ts = findInterval(data[i, ], ybins)
        for (j in 1:ncol(data)){
            hist.matrix[ts[j], j] = hist.matrix[ts[j], j] + 1
        }
    }

    hist.matrix
}

# Define the months variable
months <- month.abb

# Create the plot and save it as PNG
png(filename = "Figura3-13.png")

# Assuming 'air' and 'AirPassengers_Num' are predefined in your R environment
h = hist2d(t(matrix(air$AirPassengers_Num, nrow = 12, ncol = 12)), 5, months)
image(1:ncol(h), 1:nrow(h), t(h), col = heat.colors(5), axes = FALSE, xlab = 'Time', ylab = 'Passenger Count')

dev.off()
"""

robjects.r(cod)
```

```python
cod="""# Load required library
require(data.table)

# Define the correct URL string for the raw file
url.str <- "https://raw.githubusercontent.com/AileenNielsen/TimeSeriesAnalysisWithPython/master/data/50words_TEST.csv"

# Read the data from the URL
words <- fread(url.str)

# Inspect column names
print(colnames(words))

# Assuming 'v1' is the correct column name, otherwise use the appropriate column name
if ("V1" %in% colnames(words)) {
    w1 <- words[V1 == 1]
} else {
    stop("Column 'v1' not found. Check the column names in the dataset.")
}

# Define hist2d function as before
hist2d <- function(data, nbins.y, xlabels){
    ymin = min(data)
    ymax = max(data) * 1.0001
    ybins = seq(from = ymin, to = ymax, length.out = nbins.y + 1)
    hist.matrix = matrix(0, nrow = nbins.y, ncol = ncol(data))
    for (i in 1:nrow(data)){
        ts = findInterval(data[i, ], ybins)
        for (j in 1:ncol(data)){
            hist.matrix[ts[j], j] = hist.matrix[ts[j], j] + 1
        }
    }
    hist.matrix
}

# Create the plot and save it as PNG
png(filename = "Figura3-15.png")

h = hist2d(w1, 25, 1:ncol(w1))

colors <- gray.colors(20, start = 1, end = .5)
par(mfrow = c(1, 2))
image(1:ncol(h), 1:nrow(h), t(h), col = colors, axes= FALSE, xlab= 'Time', ylab= 'Projection Value')

image(1:ncol(h), 1:nrow(h), t(log(h + 1)), col = colors, axes= FALSE, xlab= 'Time', ylab= 'Projection Value')

dev.off()"""

robjects.r(cod)
```
